```python
import numpy as np
center=np.array(ann['translation'])#GT的中心坐标
orientation=np.array(ann['rotation'])#GT边界框的方向，四元数形式表示
print(center)
print(orientation)
```

```python
from pyquaternion import Quaternion
quaternion=Quaternion(ego_data['rotation']).inverse#生成四元数求其逆
```

```python
print(quaternion)
```

```python
center -=np.array(ego_data['translation'])#GT的中心坐标平移操作
print(center)
```

```python
center = np.dot(quaternion.rotation_matrix,center)#生成旋转矩阵，进行旋转操作
print(center)#此时GT的中心坐标即为新坐标系——车身坐标系下的坐标
```

```python
orientation=quaternion*orientation#旋转操作，将GT的边界框方向，变换为车身坐标系下
print(orientation)#此时GT的方向，变换为车身坐标系下
```
